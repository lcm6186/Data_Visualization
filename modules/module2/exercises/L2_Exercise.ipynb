**Activity 3:** Use ellipse library to create an elliptical correlation plot for the diamonds data using carat, price, table, and depth variables. Make sure to use a **colorblind safe, diverging** color scheme. 

```r
library(ellipse)
head(diamonds)
data=diamonds[,c("carat","depth","price","table")]
## Your answer for activity 3 goes here ---

## Setting up a correlation matrix
datacor = cor(data)
datacor

## Set up colors to used 
my_colors <- brewer.pal(10, "RdYlBu")
my_colors <- colorRampPalette(my_colors)(100)
my_colors <- dichromat(my_colors)
 
## Order the correlation matrix for easier viewing
ord <- order(datacor[1, ])
datacor_ord = datacor[ord, ord]

## printing the plot
plotcorr(datacor_ord , col=my_colors[datacor_ord*50+50] , mar=c(1,1,1,1)  )
```

**Activity 4:** Create a scatter plot using **all** diamonds data for carat vs. price and encode clarity variable with a **sequential** color scheme. 

```r
## Your answer for activity 4 goes here ---

ggplot(diamonds, aes(x=carat, y=price, colour=clarity)) + 
    geom_point() + 
    scale_color_brewer(palette='YlGnBu') + 
    theme_dark()

## I added the dark theme to make the lighter points easier to see
```

**Activity 5:** Repeat Activity 4 with **alpha transparency**. **Is this a better idea? Discuss your answer.**

As for the alpha being a better idea; for this plot having a lighter transparency I believe helps to more easily see what is going on in the heavier concentrated areas of the plot, particularly where carat < 1.5 and price < 5000. 

```r
## Your answer for activity 5 goes here ---

ggplot(diamonds, aes(x=carat, y=price, colour=clarity)) + 
    geom_point(alpha=0.5) + 
    scale_color_brewer(palette='YlGnBu') + 
    theme_dark()

# hint: geom_point(alpha=0.5)
```
