```r
## Setting up seperate frame to ensure I preserve the original

USDA_NA <- USDA

## Testing Code to write as function
# USDA_NA$Calories <- ifelse(is.na(USDA_NA$Calories), median(USDA_NA$Calories, na.rm=TRUE), USDA_NA$Calories)
# USDA_NA$Protein <- ifelse(is.na(USDA_NA$Protein), median(USDA_NA$Protein, na.rm=TRUE), USDA_NA$Protein)
# USDA_NA$VitaminD <- ifelse(is.na(USDA_NA$VitaminD), median(USDA_NA$VitaminD, na.rm=TRUE), USDA_NA$VitaminD)

## Testing above
# sum(is.na(USDA$Calories))
# sum(is.na(USDA_NA$Calories))
# sum(is.na(USDA$Protein))
# sum(is.na(USDA_NA$Protein))
# sum(is.na(USDA$VitaminD))
# sum(is.na(USDA_NA$VitaminD))

## Since this works we will write a function to make this easier

## Creating a function

fillColNA <- function(x) {
  ifelse(is.na(x), median(x, na.rm=TRUE), x)
}


## Testing function

# USDA_NA$VitaminE <- fillColNA(USDA_NA$VitaminE)
# sum(is.na(USDA$VitaminE))
# sum(is.na(USDA_NA$VitaminE))

# USDA_NA$VitaminC <- fillColNA(USDA_NA$VitaminC)
# sum(is.na(USDA$VitaminC))
# sum(is.na(USDA_NA$VitaminC))

## Make sure of results
# head(USDA_NA, n=10)
```

```r
## Your answer for activity 2 goes here ---

## Running the above created function through a for loop over ever column. This is more effecient than writing out
## each function call by hand for each column

for (i in colnames(USDA_NA[3:16])){
    USDA_NA[[i]] <- fillColNA(USDA_NA[[i]])
}
```

**Activity 3:** Find the correlation between the variables in the dataset and find the pairs which have correlation more than 0.5. 

*Hint: there are 4 pairs of them.*

Programmably **list them** or **create a visualization** that one can read the answer off of it easily, as opposed to just printing a correlation matrix.

```r
## Your answer for activity 3 goes here ---

USDA_COR <- cor(USDA_NA[,3:16])

for (i in 1:nrow(USDA_COR)){
  correlations <- which((USDA_COR[i,] > 0.5) & (USDA_COR[i,] != 1))
  
  if(length(correlations)> 0){
    print(colnames(USDA_COR)[i])
    print(correlations)
  }
}

## In the below out we can see the 5 variables that have correlations greater than .5 with another variable. 3 of these
## variables have correlations greater than .5 with more than one variable, which gives us the 4 pairs as:

## 1. Calories | Total Fat
## 2. Calories | Saturated Fat
## 3. Carbohydrates | Sugar
## 4. Total Fat | Saturated Fat
```
